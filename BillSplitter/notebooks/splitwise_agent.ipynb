```python
from splitwise import Splitwise
import os
from dotenv import load_dotenv

load_dotenv()

sObj = Splitwise(os.environ.get("SPLITWISE_CONSUMER_KEY"), os.environ.get("SPLITWISE_CONSUMER_SECRET"), api_key=os.environ.get("SPLITWISE_API_KEY"))

current = sObj.getCurrentUser()
print(current.getEmail()) 
```

```python
for group in sObj.getGroups():
    if "at&t" in group.getName().lower():
        group_id = group.getId()
        for member in group.getMembers():
            print(member.getFirstName(), member.getLastName(), member.getEmail())
        break
```
